```python
sandwich_orders = ["Club Sandwich", "BLT", "Grilled Cheese"]
finished_sandwichs = []

while sandwich_orders:
    current_sandwich = sandwich_orders.pop()
    print(f"\nFinishing sandwich: {current_sandwich}")
    finished_sandwichs.append(current_sandwich)

print("\nAll finished sandwiches:")
for finished_sandwich in finished_sandwichs:
    print(finished_sandwich)
```

```python
responses = {}

polling_active = True

while polling_active:
    name = input("\nWhat is your name?")
    response = input("If you could visit one place in the world, where would you go?")
    responses[name] = response

    repeat = input("\nWill anyone else like to response? (yes/no)")
    if repeat == "no":
        polling_active = False

print("\nPoll Results")
for name, response in responses.items():
    print(f"{name} would like to go to {response}.")
```
