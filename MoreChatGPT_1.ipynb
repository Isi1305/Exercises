‚≠ê √úbung 2 ‚Äî Men√ºpr√ºfung (√§hnlich wie eben)

Du hast:

Pizza = 6

Burger = 8

Water = 2

Wenn etwas anderes eingegeben wird ‚Üí Preis = 5.

üëâ Input: mehrere Items getrennt durch Leerzeichen.
üëâ Gesamtpreis berechnen.

```python
orders = input()
orders = orders.split()
order = order.lower()
total = 0

for order in orders:
    if order == "pizza":
        total += 6
    elif order == "burger":
        total += 8
    elif order == "water":
        total += 2
    else:
        total += 5
print(f"Es sind {total} ‚Ç¨")
```

‚≠ê √úbung 3 ‚Äî Durchschnitt berechnen

User gibt Zahlen ein:

5 10 15 20


üëâ Berechne den Durchschnitt.

Das brauchst du:

Summe

Anzahl der Zahlen

```python
liste = input()
liste = liste.split()
summe = 0  

for number in liste:
    summe += int(number)

average = summe / len(liste)
print(average)
```

‚≠ê √úbung 4 ‚Äî Gro√ü-/Kleinschreibung fixen (super wichtig!)

Input:

üëâ Alles soll trotzdem erkannt werden.

```python
words = input().split()

for word in words:
    if word.lower() == "pizza":
        print("Gefunden: Pizza")
    elif word.lower() == "water":
        print("Gefunden: Water")
    elif word.lower() == "burger":
        print("Gefunden: Burger")
```
