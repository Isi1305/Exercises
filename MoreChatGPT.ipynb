â­ Level 3 â€“ Data-Science-Style ğŸ˜ˆ
ğŸ‘‰ Ãœbung G â€“ Median berechnen

Input:

1 5 3 8 2


ğŸ‘‰ Sortieren
ğŸ‘‰ Mittleren Wert finden.

```python
numbers = [int(n) for n in input().split()]
numbers.sort()
if len(numbers) % 2 == 0:
    left = len(numbers) // 2 - 1
    right = len(numbers) // 2
    median = (numbers[left] + numbers[right]) / 2
    print(f"Median: {median}")
else:
    median = numbers[len(numbers) // 2]
    print(f"Median: {median}")
```

â­ Mini-Projekt (Portfolio-tauglich)
ğŸ‘‰ Ãœbung I â€“ Noten-Auswertung

User gibt Noten ein:

1 2 2 3 1 4


Programm gibt aus:

Durchschnittsnote

Beste Note

Schlechteste Note

```python
grades = [int(n) for n in input().split()]
bestanden = 0

average = sum(grades) / len(grades)
worst_grade = max(grades)
best_grade = min(grades)

print(f"Der Durchschnitt ist: {average}")
print(round(average))
print(f"Die schlechteste Note war: {worst_grade}")
print(f"Die beste Note war: {best_grade}")

for grade in grades:
    if grade <= 4:
        print(f"Mit einer {grade} bestanden!") # Das hatte ich falsch verstanden von der Aufgabe aber habe es drin gelasen.
        bestanden += 1
    else:
        print(f"Mit einer {grade} nicht bestanden!")

print(f"So viele FÃ¤cher wurden bestanden: {bestanden}")
```
